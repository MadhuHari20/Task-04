```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import matplotlib.pyplot as plt
import pandas as pd

# Sample social media data
data = {
    'text': [
        "I love this product! It's amazing.",
        "Not happy with the customer service.",
        "Neutral opinion about this brand.",
        "The new feature is really cool."
    ]
}

df = pd.DataFrame(data)

# Preprocess the data (for simplicity, we assume it's clean here)

# Initialize sentiment analyzer
nltk.download('vader_lexicon')
sid = SentimentIntensityAnalyzer()

# Perform sentiment analysis
df['sentiment'] = df['text'].apply(lambda text: sid.polarity_scores(text)['compound'])

# Visualize sentiment distribution
sentiment_counts = df['sentiment'].value_counts(normalize=True) * 100

labels = sentiment_counts.index
sizes = sentiment_counts.values

plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['lightgreen', 'lightcoral', 'lightskyblue'])
plt.title('Sentiment Distribution')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
